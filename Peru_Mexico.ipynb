**Produccion de Plata por Mes entre Perú y Mexico**

```python
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTUeoCWGNC0C1vEqaJi3VlbTL9mE62SWG44H0TKWwHb6KqoA_95f7XPGe98ZmStYQ/pub?output=csv", index_col="Fecha",)
```

```python
datos
```

```python
# Se utiliza para contar la cantidad de valores nulos en cada columna de tu DataFrame datos y luego imprimir estos recuentos.
print(datos.isnull().sum())
```

```python
# Verifica el DataFrame (datos) si alguna de ellas tiene un tipo de dato de objeto (dtype == "object").
# Si encuentra una columna con este tipo de dato, intentará convertir sus valores a tipo numérico utilizando la función pd.to_numeric().
for col in datos.columns:
    if datos[col].dtype == "object":
        try:
            datos[col] = pd.to_numeric(datos[col])
        except:
            pass
```
