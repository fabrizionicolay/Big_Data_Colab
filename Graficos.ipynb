```python
fig, graficos = plt.subplots(1,2,figsize=(8,6))

#Graficamos "puntos" en las coordenadas X e Y en el primer grafico (0)
graficos[0].scatter(
    x=[1,2,3],
    y=[3,2,1]
)
#Graficamos una funci√≥n cuadr√°tica con valores desde -10 hasta 10 de 0.25, para
#ello nos apoyamos en numpy y su funci√≥n 'arange'.
eje_x = np.arange(-10,10,0.25)
eje_y = []#aqu√≠ guardaremos los cuadrados

for numero in eje_x:
    cuadrado = numero*numero #calculamos el cuadrado de cada elemento
    eje_y.append(cuadrado) #lo agregamos a 'eje_y'

#graficos[1].scatter(
#    x = eje_x,
#    y = eje_y
#)

#graficamos con 'plot' en vez de 'scatter' para dibujar lineas en vez de puntos.
graficos[1].plot(
    eje_x,
    eje_y
)

plt.show() #visualizamos ambos gr√°ficos.
```

#GRAFICO 4

```python
#importamos la libreria matplotlib
import matplotlib.pyplot as plt
import numpy as np

#configuramos el grafico_
# indicamos que la distribucion de esos 2 graficos sera de 1 fila y 2 columnas
# Tama√±o del grafico sera de 8*6 pulgadas

fig, graficos = plt.subplots(figsize=(8,6))

eje_x = np.arange(-10,10,0.2) #valores cualquiera que para el ejemplo se espresa en Radianes
eje_y = [] #aqui guardamos el valor de la funci√≥n Seno para cada elemento de "eje_x"

for number in eje_x: #iteramos...
    vseno = np.sin(number) #calculamos el seno de cada numero de "eje_x"
    eje_y.append(vseno) #lo almacenamos
    print(vseno)

#graficamos
graficos.plot(
    eje_x,
    eje_y
)

plt.show()#visualizamos ambos gr√°ficos.
```
