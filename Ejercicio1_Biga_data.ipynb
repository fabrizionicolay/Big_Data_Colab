```python
# Obtenemos el promedio de ingresos por ciudad

import pandas as pd

# Load the data from the CSV file
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRyAsF1thWCw3xEJsw9K2Fg6P-nKDaYxwYK9kOfLKnjkV2AMoTZORNKjouNq-TxCsLDZ38ybGrO8VYM/pub?gid=2028458572&single=true&output=csv")

# Calculate the average income per city
average_income_by_city = datos.groupby('ciudad')['ingresos'].mean().astype(int)

# Print the average income per city
print(average_income_by_city)
```

```python
#Obtenemos la persona mas joven y mas mayor

persona_joven = datos.loc[datos['edad'] == datos['edad'].min(), ['nombre', 'edad']]
persona_mayor = datos.loc[datos['edad'] == datos['edad'].max(), ['nombre', 'edad']]

print("Persona mas Joven:", persona_joven.values[0])
print("Persona mas Mayor:", persona_mayor.values[0])
```

```python
#Obtenemos el total de trabajadores por ciudad
personas_ciudad = datos['ciudad'].value_counts()

print(personas_ciudad)
```

```python
#Obtenemos el orden por edad de los personas
#datos.sort_values(by=['edad'], ascending=False)[['nombre', 'edad']]
datos.sort_values(by=['edad'], ascending=False)
```

```python
#Obtenemos el promedio de ingresos de personas que son mayores de 30 años

datos_mayores_30 = datos[datos['edad'] > 30]
promedio_ingresos_mayores_30 = int(datos_mayores_30['ingresos'].mean())
print("El promedio de ingresos de personas mayores de 30 años es:", promedio_ingresos_mayores_30)
```

```python
# Obtenemos los ingresos totales por ciudades

ingresos_totales_por_ciudad = datos.groupby('ciudad')['ingresos'].sum()
print(ingresos_totales_por_ciudad)
```

```python
#Obtenemos la persona con el ingreso mas alto entre todas las ciudades

persona_mayor_ingreso = datos.loc[datos['ingresos'] == datos['ingresos'].max(), ['nombre', 'ingresos']]
print("Persona con mayor ingreso:", persona_mayor_ingreso.values[0])
```

```python
# Obtenemos un histograma que representa las edades de los trabajadores en la ciudad

import matplotlib.pyplot as plt

# Crea un histograma de las edades
plt.hist(datos['edad'], bins=5)

# Agregar etiquetas y título
plt.xlabel("Edades")
plt.ylabel("Numero de Personas")
plt.title("Histograma de edades")

plt.show()
```

```python
#Obtenemos una columna de impuestos que represente el 10% de los ingresos de todos los trabajadores

datos['impuestos'] = datos['ingresos'] * 0.1
datos.head()
```
